```python
# Atividade 1

# Exemplo usando arange
arr_arange = np.arange(0, 10)
print("Array com arange:", arr_arange)

# Exemplo usando linspace
arr_linspace = np.linspace(0, 100, 15)
print("Array com linspace:", arr_linspace)
```

```python
# Atividade 2

# Exemplo de power() e sqrt()
random_integers = np.random.randint(1, 11, 5)
print("Array original:", random_integers)

# Power (elevando ao quadrado)
squared = np.power(random_integers, 2)
print("Quadrados dos números:", squared)

# Sqrt (raiz quadrada)
sqrt_values = np.sqrt(random_integers)
print("Raízes quadradas:", sqrt_values)
```

```python
# Atividade 3

# Criando o array
arr = np.linspace(0, 5, 10)
print("Array gerado:", arr)

# Power (elevando ao cubo)
cubed = np.power(arr, 3)
print("Potência de 3:", cubed)

# Sqrt (raiz quadrada)
sqrt_arr = np.sqrt(arr)
print("Raízes quadradas:", sqrt_arr)
```
