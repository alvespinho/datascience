```python
# Criando o conjunto de dominó
domino = [(i, j) for i in range(7) for j in range(i, 7)]  # Todas as 28 peças

# Evento: peças duplas
buchas = [peca for peca in domino if peca[0] == peca[1]]

# Espaço amostral
total_pecas = len(domino)  # Ω = 28

# Probabilidade
p_dupla = len(buchas) / total_pecas
print(f"Probabilidade de sortear uma bucha: {p_dupla:.2f}")

# Mostrando quais são as peças do evento
print("Buchas:", buchas)
```
