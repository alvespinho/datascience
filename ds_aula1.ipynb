```python
'''
      1. Você foi contratado para registrar cidades e suas
            respectivas temperaturas:

           Salvador (28.5), São Paulo (21.3), Fortaleza (25.4),
           Recife (30.1), Belo Horizonte (29.3)

          Resolva usando o método Series do Pandas.

'''
```

```python
cidades = ["Salvador", "São Paulo", "Fortaleza", "Recife", "Belo Horizonte"]
temperaturas = [28.5, 21.3, 25.4, 30.1, 29.3]

serie_temperaturas = pd.Series(temperaturas, index= cidades)
serie_temperaturas
```

```python
'''
      2. A partir da Series criada, responda:

          a) Qual a temperatura registrada em Recife?
          b) Qual foi a temperatura média entre as 5 cidades?
          c) Quais cidades estão com temperatura acima de 26°C?
          d) Substitua a temperatura de São Paulo para 22.0.
'''
```

```python
#a)
print(serie_temperaturas["Recife"])

#b)
print(serie_temperaturas.mean())

#c)
print(serie_temperaturas[serie_temperaturas > 26])

#d)
serie_temperaturas["São Paulo"] = 22.0
print(serie_temperaturas)
```
