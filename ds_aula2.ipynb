```python
df_partidas['date'] = pd.to_datetime(df_partidas['date'], format= '%Y-%m-%d %H:%M')
df_partidas
```

```python
# Cria coluna com o dia apenas para exibição
df_partidas['dia'] = df_partidas['date'].dt.day
df_partidas['mes'] = df_partidas['date'].dt.month

# Agrupa por dia, mas mantém a ordem de aparição original
contagem = df_partidas.groupby(['mes', 'dia'], sort=False)['date'].count()

# # Conta o número de partidas por mês e dia, sem alterar a ordem original
# contagem = df_partidas.groupby(['mes', 'dia'], sort=False)['date'].count().reset_index(name='n_partidas')

print(contagem)
```

```python
memoria_antes = df_partidas.memory_usage(deep=True).sum()
print (memoria_antes)
```

```python
# Verifica os tipos antes
print("Antes do downcast:")
print(df_partidas.dtypes.value_counts())
```

```python
# Seleciona apenas colunas do tipo int64
colunas_int64 = df_partidas.select_dtypes(include=['int64']).columns

# Aplica downcast para inteiro menor possível (int32, int16 ou int8)
for col in colunas_int64:
    df_partidas[col] = pd.to_numeric(df_partidas[col], downcast='integer')

# Verifica os tipos depois
print("\nDepois do downcast:")
print(df_partidas.dtypes)
```

```python
memoria_depois = df_partidas.memory_usage(deep=True).sum()

print(f"Memória antes: {memoria_antes / 1024:.2f} KB")
print(f"Memória depois: {memoria_depois / 1024:.2f} KB")
print(f"Economia: {100 * (memoria_antes - memoria_depois) / memoria_antes:.2f}%")
```
